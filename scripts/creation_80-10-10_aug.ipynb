```python
import os
import random

images_dir = "export_michelin_v1/images"
output_dir = "data_aug"
os.makedirs(output_dir, exist_ok=True)

# list images (without extension)
image_names = [
    os.path.splitext(f)[0]
    for f in os.listdir(images_dir)
    if f.lower().endswith((".jpg", ".png"))
]

image_names.sort()
random.shuffle(image_names)

# split (80 / 10 / 10)
n = len(image_names)
train_split = int(0.8 * n)
val_split   = int(0.9 * n)

train = image_names[:train_split]
val   = image_names[train_split:val_split]
test  = image_names[val_split:]
total = len(train) + len(val) + len(test)

def save_list(path, names):
    with open(path, "w") as f:
        for n in names:
            f.write(n + "\n")

save_list(os.path.join(output_dir, "image_names_train.txt"), train)
save_list(os.path.join(output_dir, "image_names_val.txt"), val)
save_list(os.path.join(output_dir, "image_names_test.txt"), test)


print("Files created in data_aug/:")
print(f"total: {total} images")
print(f" train: {len(train)} images")
print(f" val:   {len(val)} images")
print(f" test:  {len(test)} images")
```
